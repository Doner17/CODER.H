```python
import pandas as pd
import glob
import os

# Lista de archivos CSV y Excel
file_patterns = ["/content/drive/MyDrive/Data Science-CODER/DESAFIOS/Desafío de visualización /VARIABLES/005930.KS.csv",
                 "/content/drive/MyDrive/Data Science-CODER/DESAFIOS/Desafío de visualización /VARIABLES/AAPL.csv",
                 "/content/drive/MyDrive/Data Science-CODER/DESAFIOS/Desafío de visualización /VARIABLES/MSFT.csv",
                 "/content/drive/MyDrive/Data Science-CODER/DESAFIOS/Desafío de visualización /VARIABLES/GOOG.csv",
                 "/content/drive/MyDrive/Data Science-CODER/DESAFIOS/Desafío de visualización /VARIABLES/^VIX.csv",
                 "/content/drive/MyDrive/Data Science-CODER/DESAFIOS/Desafío de visualización /VARIABLES/Estados Unidos 2 años Información Histórica Rendimiento Bono.csv",
                 "/content/drive/MyDrive/Data Science-CODER/DESAFIOS/Desafío de visualización /VARIABLES/Estados Unidos 30 años Datos Históricos Rendimiento de Bonos.csv"
                 ]
# Leer cada archivo y cargarlo en un dataframe
dfs = []
for pattern in file_patterns:
    file_paths = glob.glob(pattern)
    for file_path in file_paths:
        if file_path.endswith('.csv'):
            df = pd.read_csv(file_path)
            if 'Date' in df.columns:
                df.rename(columns={'Date': 'Fecha'}, inplace=True)
            df['Fecha'] = pd.to_datetime(df['Fecha'])
        elif file_path.endswith('.xlsx') or file_path.endswith('.xls'):
            df = pd.read_excel(file_path)
            if 'Fecha' not in df.columns:
                continue
            df['Fecha'] = pd.to_datetime(df['Fecha'], format='%d.%m.%Y')
        dfs.append(df)

# Combinar los dataframes en uno solo basado en las fechas
df_combinado = dfs[0]
for i in range(1, len(dfs)):
    df_combinado = pd.merge(df_combinado, dfs[i], on='Fecha', how='inner')

# Ruta y nombre del archivo de destino
output_folder = '/content'
file_name = 'Apple y sus variable8.xlsx'

# Construir la ruta completa del archivo
output_path = os.path.join(output_folder, file_name)

# Guardar el archivo de Excel temporalmente
df_combinado.to_excel(output_path, index=False)
```

```python
import pandas as pd
import glob
import os

# Lista de archivos CSV y Excel
file_patterns = [
    "/content/drive/MyDrive/Data Science-CODER/DESAFIOS/Desafío de visualización /VARIABLES/005930.KS.csv",
    "/content/drive/MyDrive/Data Science-CODER/DESAFIOS/Desafío de visualización /VARIABLES/AAPL.csv",
    "/content/drive/MyDrive/Data Science-CODER/DESAFIOS/Desafío de visualización /VARIABLES/GOOG.csv",
    "/content/drive/MyDrive/Data Science-CODER/DESAFIOS/Desafío de visualización /VARIABLES/MSFT.csv",
    "/content/drive/MyDrive/Data Science-CODER/DESAFIOS/Desafío de visualización /VARIABLES/^VIX.csv",
    "/content/drive/MyDrive/Data Science-CODER/DESAFIOS/Desafío de visualización /VARIABLES/Estados Unidos 2 años Información Histórica Rendimiento Bono.csv",
    "/content/drive/MyDrive/Data Science-CODER/DESAFIOS/Desafío de visualización /VARIABLES/Estados Unidos 30 años Datos Históricos Rendimiento de Bonos.csv"
]

# Leer cada archivo y cargarlo en un DataFrame
dfs = []
for pattern in file_patterns:
    file_paths = glob.glob(pattern)
    for file_path in file_paths:
        if file_path.endswith('.csv'):
            df = pd.read_csv(file_path)
            if 'Date' in df.columns:
                df.rename(columns={'Date': 'Fecha'}, inplace=True)
            df['Fecha'] = pd.to_datetime(df['Fecha'])
        elif file_path.endswith('.xlsx') or file_path.endswith('.xls'):
            df = pd.read_excel(file_path)
            if 'Fecha' not in df.columns:
                continue
            df['Fecha'] = pd.to_datetime(df['Fecha'], format='%d.%m.%Y')
        dfs.append(df)

# Combinar los DataFrames en uno solo
df_combinado = dfs[0]
for i in range(1, len(dfs)):
    df_combinado = pd.merge(df_combinado, dfs[i], on='Fecha', how='inner')

# Eliminar la columna número 31
df_combinado = df_combinado.drop(columns=[df_combinado.columns[30]])

# Cambiar los nombres de las columnas según el orden proporcionado
new_column_names = [
    "Fecha", "Open_s", "High_s", "Low_s", "Close_s", "Adj Close_s", "Volume_s",
    "Open_a", "High_a", "Low_a", "Close_a", "Adj Close_a", "Volume_a",
    "Open_m", "High_m", "Low_m", "Close_m", "Adj Close_m", "Volume_m",
    "Open_g", "High_g", "Low_g", "Close_g", "Adj Close_g", "Volume_g",
    "Open_v", "High_v", "Low_v", "Close_v", "Adj Close_v",
    "Close_2", "Open_2", "High_2", "Low_2", "% var_2",
    "Close_30", "Open_30", "High_30", "Low_30", "% var_30",
]

# Cambiar los nombres de las columnas en el DataFrame
df_combinado.columns = new_column_names

# Modificar el formato de las columnas % var_2 y % var_30
df_combinado['% var_2'] = df_combinado['% var_2'].str.replace(',', '.').str.rstrip('%').astype(float) / 100
df_combinado['% var_30'] = df_combinado['% var_30'].str.replace(',', '.').str.rstrip('%').astype(float) / 100

# Ruta y nombre del archivo de destino
output_folder = '/content'
file_name = 'Apple y sus variable2.xlsx'

# Construir la ruta completa del archivo
output_path = os.path.join(output_folder, file_name)

# Guardar el archivo de Excel
df_combinado.to_excel(output_path, index=False)

# Imprimir la ruta del archivo
print("El archivo se ha guardado en:", output_path)

df_combinado = pd.merge(df_combinado, dfs[i], on='Fecha', how='inner', suffixes=('', ''))
```

```python
import yfinance as yf
import pandas as pd

# Definición del símbolo de la acción
symbol = "INTC"

# Define las fechas de inicio y finalización del historial
start_date = "2013-07-15"
end_date = "2023-07-15"

# Crea un objeto Ticker para el símbolo especificado
ticker = yf.Ticker(symbol)

# Obtiene el historial de precios en el rango de fechas especificado
historical_data = ticker.history(period="1d", start=start_date, end=end_date)

# Reinicia el índice y luego cambia el nombre de la columna de índice a "Fecha"
historical_data = historical_data.reset_index()
historical_data.rename(columns={'Date': 'Fecha'}, inplace=True)

# Ajustar el formato de la columna "Fecha" eliminando la hora
historical_data['Fecha'] = historical_data['Fecha'].dt.strftime('%Y-%m-%d')

# Cambia los nombres de las columnas
historical_data.rename(columns={'Open': 'Open_i', 'High': 'High_i', 'Low': 'Low_i', 'Close': 'Close_i', 'Volume': 'Volume_i'}, inplace=True)

# Elimina las columnas "Dividends" y "Stock Splits" del DataFrame
historical_data = historical_data.drop(["Dividends", "Stock Splits"], axis=1)
```

```python
import numpy as np
from scipy import stats

# Extrae data
X = df_combinado["Close_a"].values
S = df_combinado["Close_s"].values

# Calcula el coeficiente de correlación de Pearson y el p-valor
correlation, p_value = stats.pearsonr(X, S)

# Establece el umbral deseado para la correlación
umbral_correlacion = 0.6
alfa = 0.05  # Nivel de significancia

# Imprime los valores de correlación y p-valor
print("Coeficiente de correlación de Pearson:", correlation)
print("Valor p:", p_value)

# Compara la correlación con el umbral y el p-valor con el nivel de significancia
if correlation > umbral_correlacion and p_value < alfa:
    print("La correlación entre Apple y Samsung es mayor que 0.6 y es estadísticamente significativa. Se rechaza la hipótesis nula.")
elif correlation > umbral_correlacion:
    print("La correlación entre Apple y Samsung es mayor que 0.6, pero no es estadísticamente significativa. No se rechaza la hipótesis nula.")
else:
    print("La correlación entre Apple y Samsung no es mayor que 0.6 o no es estadísticamente significativa. No se rechaza la hipótesis nula.")
```

```python
import pandas as pd
from scipy.stats import ttest_rel
from IPython.display import HTML

# Realizar las pruebas t de Student para muestras emparejadas
t_statistic_am, p_value_am = ttest_rel(df_combinado['Close_a'], df_combinado['Close_m'])
t_statistic_ai, p_value_ai = ttest_rel(df_combinado['Close_a'], df_combinado['Close_i'])
t_statistic_ag, p_value_ag = ttest_rel(df_combinado['Close_a'], df_combinado['Close_g'])

# Interpretar los resultados
alpha = 0.05  # Nivel de significancia

# Crear una lista de diccionarios con los resultados
resultados = [
    {
        "Variable": "Apple vs. Microsoft",
        "Estadístico t": t_statistic_am,
        "Valor p": p_value_am,
        "Resultado": "Los precios de cierre de Apple son significativamente mayores que los de Microsoft." if p_value_am < alpha else "No se puede concluir que los precios de cierre de Apple sean significativamente mayores que los de Microsoft."
    },
    {
        "Variable": "Apple vs. Intel",
        "Estadístico t": t_statistic_ai,
        "Valor p": p_value_ai,
        "Resultado": "Los precios de cierre de Apple son significativamente mayores que los de Intel." if p_value_ai < alpha else "No se puede concluir que los precios de cierre de Apple sean significativamente mayores que los de Intel."
    },
    {
        "Variable": "Apple vs. Google",
        "Estadístico t": t_statistic_ag,
        "Valor p": p_value_ag,
        "Resultado": "Los precios de cierre de Apple son significativamente mayores que los de Google." if p_value_ag < alpha else "No se puede concluir que los precios de cierre de Apple sean significativamente mayores que los de Google."
    }
]

# Crear un DataFrame a partir de los resultados
df_resultados = pd.DataFrame(resultados)

# Convertir el DataFrame a una tabla HTML
html_table = df_resultados.to_html(index=False, escape=False, classes='table table-striped table-bordered')

# Formatear el código HTML con el estilo proporcionado
styled_html = f'''
<div style="max-width: 1200px; margin: 0 auto; text-align: center; border: 2px solid #3f5edf; border-radius: 10px; background-color: #f1f3fd; padding: 10px; color: black;">
  <h2 style="color: #3f5edf; font-weight: bold; font-size: 12px;">Resumen del test de hipótesis sobre los precios de cierre de Apple</h2>
  {html_table.replace('<th>Variable</th>', '<th style="color: #3f5edf;">Variable</th>').replace('<th>Estadístico t</th>', '<th style="color: #3f5edf;">Estadístico t</th>').replace('<th>Valor p</th>', '<th style="color: #3f5edf;">Valor p</th>').replace('<th>Resultado</th>', '<th style="color: #3f5edf;">Resultado</th>')}
</div>
'''

# Mostrar la tabla HTML formateada en Colab
display(HTML(styled_html))
```
